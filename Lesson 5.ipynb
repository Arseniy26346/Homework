```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5, 0.001)
y = 3*x**2
data = pd.DataFrame({'x': x, 'y': y})
```

```python
sns.lineplot(data = data, y = 'y', x = 'x')
```

```python
x0=5
dfx = 6*x
x0 = 5
l = 0.1
x1 = x0 - l*dfx
res = [x1]
```

```python
x0=10
dfx = 6*x0
l = 0.1
x1 = x0 - l*dfx
res = [x1]
while abs(x0-x1)> 0.000000001:
    x0 = x1
    dfx = 6*x0
    x1 = x0 - l*dfx
    res.append(x1)
print(x1)
```

```python
len(res)
```

```python
df = pd.DataFrame({'x' : res})
df['y'] = 3*df['x']**2
df
```

```python
sns.lineplot(data = df, y = 'y', x = 'x')
```

```python
plt.figure(figsize = (10,6))
sns.lineplot(data = data, y = 'y', x = 'x', label = 'orig')
sns.lineplot(data = df, y = 'y', x = 'x', label = 'grad')
plt.title('  ')
plt.show()
```
